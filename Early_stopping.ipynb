```python
from sklearn.datasets import load_diabetes
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
es=EarlyStopping(monitor='loss',min_delta=0.001,patience=3,mode='min')
ld=load_diabetes()
x=ld.data
y=ld.target
ss=StandardScaler()
x1=ss.fit_transform(x)
xtrain,xtest,ytrain,ytest=train_test_split(x1,y,test_size=0.2)
model=Sequential()
model.add(Dense(12,activation='relu',input_dim=10))
model.add(Dense(12,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='relu'))
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy']) # can also define epos....
```

```python
hs=model.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=500,callbacks=[es])
```

```python
hs.history.keys()
```

```python
import matplotlib.pyplot as plt
plt.plot(hs.history['val_loss'])
plt.plot(hs.history['loss'])
plt.plot(hs.history['accuracy'])
```
