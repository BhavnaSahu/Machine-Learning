```python
import pandas as pd
df=pd.read_excel(uploaded['plot price dataset.xlsx'])
x=df['plot Price']
y=df['plot size']
len(x)
x
```

```python

class Mylr:
  def fit(self,x,y,epoch,learningrate):
    m=0
    c=0
    length=len(x)
    for j in range(epoch):
        for i in range(length):
          model_output=m*x[i]+c
          error=model_output-y[i]
          gm=-2*error*x[i]/length
          gc=-2*error/length
          m=m+(learningrate*gm)
          c=c+(learningrate*gc)

    return m,c 

  def predict(self,m,c,x):
    output=[]
    for i in range(len(x)):
      o=m*x[i]+c 
      output.append(o)
    return output
```

```python
model=Mylr()
m,c=model.fit(x,y,100,0.001)
```

```python
OT=model.predict(m,c,x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
#plt.scatter(x,OT,c='red')
plt.plot(x,OT,c='red')
```

```python
import numpy as np

xarr=np.array(x)
yarr=np.array(y)
xnew=xarr.reshape(-1,1)
ynew=yarr.reshape(-1,1)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xnew,ynew)
```

```python
prediction =lr.predict(xnew)
```

```python
plt.scatter(x,y)
plt.plot(x,prediction)
```

```python
from sklearn.metrics import r2_score # regression ki accuracy ko check karta hai..
#from sklearn.metrics import preci
r2_score(ynew,prediction)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xnew,ynew,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
model_output=lr.predict(xtest)
from sklearn.metrics import r2_score
r2_score(ytest,model_output)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xtrain,ytrain)
plt.plot(xtest,model_output)
```
