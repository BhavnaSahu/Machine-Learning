```python
#training or testing ki accuracy mai jayada difference nhi hoga...after performing this..
from sklearn.datasets import load_breast_cancer
dataset=load_breast_cancer()
x=dataset.data
y=dataset.target
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression() #learning algorithm
from sklearn.ensemble import BaggingClassifier
BG=BaggingClassifier(base_estimator=model,n_estimators=100,max_features=10,max_samples=100)
```

```python
BG.fit(xtrain,ytrain)#training
```

```python
ypred_test=BG.predict(xtest)
ypred_test
```

```python
ypred_train=BG.predict(xtrain)
ypred_train
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,ypred_test)
```

```python
accuracy_score(ytrain,ypred_train)
```

```python
#without begging classifier
model.fit(xtrain,ytrain)
ypred_test=model.predict(xtest)
ypred_train=model.predict(xtrain)
print(accuracy_score(ytest,ypred_test))
print(accuracy_score(ytrain,ypred_train))
```
